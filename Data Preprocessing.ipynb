## categorical and numerical seperations 

```python
categorical = [col for col in data.columns if data[col].dtype == 'object']

numerical = data.drop(categorical, axis=1).columns
```

```python
print(f'len of data: {len(data.columns)}, len of categorical: {len(categorical)}, len of numerical {len(numerical)}')
```

```python
# unique values in categorical values
[print(f'{col} : {data[col].unique()}') for col in categorical]
```

```python
data[categorical].isnull().mean()
```

```python
for col in numerical:
    null_ratio = data[col].isnull().mean()
    if null_ratio > 0:
        print(f"{col}: {null_ratio:.2%}") 
```

```python
# cleaning null values 

for col in categorical:
    data[col].fillna(data[col].mode()[0], inplace=True)

for col in numerical: 
    if data[col].nunique() >15:
        data[col].fillna(data[col].mean(), inplace=True)
    else:
        data[col].fillna(data[col].mode()[0], inplace=True)
    
```

```python
# Label Encoding categorical variables

for col in categorical:
    data[col] = label.fit_transform(data[col])
    print(f'{col} encoded as label form')
```

```python
data[numerical].isnull().any()
```

```python
## droping outliers

outliers = [
    'AMT_INCOME_TOTAL', 'AMT_CREDIT_x', 'AMT_ANNUITY_x', 'AMT_GOODS_PRICE_x',
    'AMT_ANNUITY_y', 'AMT_APPLICATION', 'AMT_CREDIT_y', 'AMT_DOWN_PAYMENT', 'AMT_GOODS_PRICE_y',
    'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3',
    'DAYS_BIRTH', 'DAYS_EMPLOYED', 'DAYS_REGISTRATION', 'DAYS_ID_PUBLISH',
    'DAYS_LAST_PHONE_CHANGE', 'DAYS_DECISION', 'DAYS_FIRST_DRAWING', 'DAYS_FIRST_DUE',
    'DAYS_LAST_DUE_1ST_VERSION', 'DAYS_LAST_DUE', 'DAYS_TERMINATION',
    'REGION_POPULATION_RELATIVE',
    'APARTMENTS_AVG', 'BASEMENTAREA_AVG', 'COMMONAREA_AVG', 'ELEVATORS_AVG',
    'ENTRANCES_AVG', 'FLOORSMAX_AVG', 'FLOORSMIN_AVG', 'LANDAREA_AVG',
    'LIVINGAREA_AVG', 'NONLIVINGAREA_AVG',
    'APARTMENTS_MODE', 'BASEMENTAREA_MODE', 'COMMONAREA_MODE', 'ELEVATORS_MODE',
    'ENTRANCES_MODE', 'FLOORSMAX_MODE', 'FLOORSMIN_MODE', 'LANDAREA_MODE',
    'LIVINGAREA_MODE', 'NONLIVINGAREA_MODE',
    'APARTMENTS_MEDI', 'BASEMENTAREA_MEDI', 'COMMONAREA_MEDI', 'ELEVATORS_MEDI',
    'ENTRANCES_MEDI', 'FLOORSMAX_MEDI', 'FLOORSMIN_MEDI', 'LANDAREA_MEDI',
    'LIVINGAREA_MEDI', 'NONLIVINGAREA_MEDI', 'TOTALAREA_MODE'
]
for i in outliers:
    if i != 'TARGET':
        q1 = data[i].quantile(0.25)
        q3 = data[i].quantile(0.75)
        iqr = q3 - q1
    
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
    
        data = data[(data[i] >= lower_bound) & (data[i] <= upper_bound)]

print('outliers dropped')
```
