## Exploratory data Analytics - loan data

```python
# importing required packages 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cl_data = pd.read_csv('important_data.csv')

cl_data = cl_data.iloc[:, 1:]
```

```python
# reading the data 

raw_data = pd.read_csv('loan_data.csv')


raw_data = raw_data[cl_data.columns]
```

```python
cl_data['TARGET'].value_counts()
```

```python
# categorical and numerical split

categorical = [col for col in raw_data.columns if raw_data[col].dtype == 'object']

numerical = cl_data.drop(categorical, axis=1).columns
```

```python
categorical
```

## vizualizations

```python
# vizualizing categorical features

fig, ax = plt.subplots(2, 4, figsize=(20, 8))
ax = ax.flatten()
for i, feature in enumerate(categorical):
    ax[i].hist(cl_data[feature])
    ax[i].set_title(feature)

plt.tight_layout()
plt.show()
```

```python
num_features = len(numerical) 

fig, ax = plt.subplots(6, 4, figsize=(20, 6 * 4))  

ax = ax.flatten()

for i, feature in enumerate(numerical):
    sns.kdeplot(data=cl_data, x=feature, ax=ax[i], fill=True) 
    ax[i].set_title(feature)


plt.tight_layout()
plt.show()
```

```python
# box plots 

num_features = len(cl_data) 

fig, ax = plt.subplots(8, 4, figsize=(20, 6 * 4))  

ax = ax.flatten()

for i, feature in enumerate(cl_data.columns):
    sns.boxplot(data=cl_data, x=feature, ax=ax[i], fill=True) 
    ax[i].set_title(feature)


plt.tight_layout()
plt.show()
```

```python
# comparison plot 

for feature in categorical:
    sns.violinplot(data=cl_data, x='TARGET', y=feature)
    plt.show()
```

```python
# ext_source comparison


for i in ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']:
    sns.kdeplot(data=cl_data, x=i)

plt.show()
```

```python
## correlation heatmap

plt.figure(figsize=(30, 25))

data_corr = cl_data.corr()
sns.heatmap(data= data_corr, cmap='coolwarm', annot=True)

plt.show()
```
